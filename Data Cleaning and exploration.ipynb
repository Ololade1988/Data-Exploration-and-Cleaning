# Data Cleaning

```python
# fill the missing value in age with median
train["Age"].median()
```

```python
age_mea=train["Age"].median()
train["Age"].fillna(age_mea, inplace=True)
```

```python
# fill the missing value in cabin with mode
cabin_mode=train["Cabin"].mode()[0]
train["Cabin"].fillna(cabin_mode,inplace=True)
```

```python
# fill the missing value in Embarked with mode
Embarked_mode=train["Embarked"].mode()[0]
train["Embarked"].fillna(Embarked_mode,inplace=True)
```

```python
# drop all the duplicates
train.drop_duplicates(inplace=True)
```

```python
# Create a new column title in the reain
t=[]
for i in train["Name"]:
    t.append(i.split()[1].strip(".").strip(","))
train["Title"]=t
```

```python
train["Title"].value_counts()
```

```python
major_titles=["Mr", "Miss", "Mrs", "Dr", "Rev", "Master"]
for i in train["Title"]:
    if i not in major_titles:
        train["Title"].replace(i,"other", inplace=True)
```

```python
u=pd.crosstab(columns=train["Survived"], index=train["Pclass"], values=train["Pclass"], aggfunc="count")
u.plot(kind="bar")
plt.title("Count of Survived", color="red")
plt.ylabel("Count")
plt.xlabel("Title")
plt.xticks(rotation=35,color="green")
plt.yticks(rotation=35,color="green")
plt.show()
```

```python
def barplot(col):
    u=pd.crosstab(columns=train["Survived"], index=train[col], values=train[col], aggfunc="count")
    u.plot(kind="bar")
    plt.title(f"Count of Survived by{col}", color="red")
    plt.ylabel("Count")
    plt.xlabel("Title")
    plt.xticks(rotation=35,color="green")
    plt.yticks(rotation=35,color="green")
```
